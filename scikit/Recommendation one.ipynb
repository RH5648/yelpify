```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

total_sum = 0 
df = pd.read_csv('chunk0.csv')
df
```

```python
features = ["stars", "text"]

def combine_features(row):
    return str(row["stars"]) + " " + row["text"]

for feature in features:
    df[feature] = df[feature].fillna('')
    
df["combined_features"] = df.apply(combine_features,axis=1)
df
```

```python
cv = CountVectorizer() 
count_matrix = cv.fit_transform(df["combined_features"])
cosine_sim = cosine_similarity(count_matrix)
print(cosine_sim)
print(cosine_sim.shape)
```

```python
def get_text_from_UID(user_id):
    return df[df.user_id == user_id]["text"].values[0]
def get_BID_from_index(index):
    return df['business_id'][index]
def xget_index_from_BID(business_id):
    return df[df['business_id'] == business_id].index[0]
```
